#### What's the mean salary by department. 
##### Tip : There are 2 ways to do it - one way is with and the other way is without pivot tables. Do it both ways.

#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title

#### We want to know the information about the mean salary and number of employees in each department and for each job title

```python
# your answer here
employee_dinamica1 = pd.pivot_table(employee,
                                        values=('Salary','Name'),
                                        index=('Department','Title'),
                                        aggfunc={'Salary': 'mean', 'Name': 'count'})
employee_dinamica1
```

#### We want to know for each department and for each job title the Salary's median,minimum,maximum and standard deviation. Comment your results

```python
# your answer here
employee_dinamica2 = pd.pivot_table(employee,
                                        values='Salary',
                                        index=('Department','Title'),
                                        aggfunc={'Salary': [np.median, min, max, np.std]})
employee_dinamica2
```

#### Based on your comments, fill the missing values with an appropriate value

```python
# your answer here
employee_dinamica2 = pd.pivot_table(employee,
                                        values='Salary',
                                        index=('Department','Title'),
                                        aggfunc={'Salary': [np.median, min, max, np.std]},
                                        fill_value=0)
employee_dinamica2
```

#### The stake holders want to know for each department the number of employees and how much money is spend on salaries. Could you provide that information? 

```python
# your answer here
employee_dinamica3 = pd.pivot_table(employee,
                                        values=('Salary', 'Name'),
                                        index='Department',
                                        aggfunc={'Salary': 'sum', 'Name': 'count'},
                                        margins=True,
                                        margins_name="Total")
employee_dinamica3
```

#### For each Department and Title, we want to know the the total years of work experience and the mean salary

```python
# your answer here
employee_dinamica4 = pd.pivot_table(employee,
                                        values=('Salary', 'Years'),
                                        index=('Department','Title'),
                                        aggfunc={'Salary': 'sum', 'Years': 'count'},
                                        margins=True,
                                        margins_name="Total")
employee_dinamica4
```

#### Show visually the frequency distribution of the time_grades dataset.

```python
# your answer here
mask = np.zeros_like(time_grades.corr(), dtype=bool)

mask[np.triu_indices_from(mask)] = True

plt.subplots(figsize=(15, 10))
cmap = sns.diverging_palette(6, 255, as_cmap=True)

figure = sns.heatmap(time_grades.corr(), mask=mask, cmap=cmap, center=0, linewidths=0.5, 
                     annot=True, fmt='.2f')
```
